# Case Study on ANOVA
XYZ Company has offices in four different zones. The company wishes to investigate the following :

● The mean sales generated by each zone.

● Total sales generated by all the zones for each month.

● Check whether all the zones generate the same amount of sales.

Help the company to carry out their study with the help of data provided.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv('Sales_data_zone_wise.csv')
df
```

```python
df.info()
```

```python
df.columns
```

# The mean sales generated by each zone

```python
sales_mean=df[['Zone - A', 'Zone - B', 'Zone - C', 'Zone - D']].mean()
sales_mean
```

```python
sales_mean.plot(kind ='bar')
plt.title("Mean sales of each zone")
plt.ylabel('Mean sales')
```

# Total sales generated by all the zones for each month

```python
total_sale = df['Zone - A']+df['Zone - B']+df['Zone - C']+df['Zone - D']
df['Total Monthly Sale']= total_sale
df
```

```python
plt.figure(figsize=(12,8))
sns.barplot(data =df ,x = "Month", y = "Total Monthly Sale")
plt.title("Total Sales by Month", fontsize = 15)
plt.xlabel("Months", fontsize=12)
plt.ylabel("Total Monthly Sale", fontsize = 10)
plt.xticks(rotation = 90)
plt.grid()
```

# Check whether all the zones generate the same amount of sales.

Null Hypothesis, Ho : There is no significant difference among the zone wise sales.
    
Alternate Hypothesis, Ha : There is  significant difference among the zone wise sales.

```python
from scipy.stats import f_oneway

fvalue, pvalue = f_oneway(df['Zone - A'], df['Zone - B'], df['Zone - C'],df['Zone - D'])
print("The F-statistic value is : " ,fvalue)
print("The Calculated p value is : " ,pvalue)

alpha = 0.05
if pvalue < alpha:
    print('Rejecting Ho')
else:
    print ('Failing to reject Ho')
```

p value is less than alpha, so we can conclude that there is significant difference among the zone wise sales
